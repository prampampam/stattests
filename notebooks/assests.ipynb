```python
%load_ext autoreload
%autoreload 2

import numpy as np
from matplotlib import pyplot as plt

import sys
sys.path.append('..')

from stattests.tests import *
from stattests.generation import generate_data
from stattests.utils import plot_cdf, plot_summary, codenames2titles,  frame_from_params, plot_from_params, save_gif_and_show
from stattests.data import rpv, apply_all_tests
from tqdm import tqdm

import seaborn as sns
sns.set_style("whitegrid")

import warnings
warnings.filterwarnings("ignore")
```

# Params

```python
success_rate = 0.02
uplift = 0.2
N = 5000
NN = 2000
beta = 1000
skew = 1
```

```python
skew_params = []
for s in np.linspace(0.1, 4, 20):
    skew_params.append({'success_rate': success_rate, 'uplift': uplift, 'beta': beta, 'skew': s, 'N': N, 'NN': NN})
```

```python
beta_params = []
for b in np.logspace(0, 3, 20)[::-1]:
    beta_params.append({'success_rate': success_rate, 'uplift': uplift, 'beta': b, 'skew': skew, 'N': N, 'NN': NN})
```

```python
sr_params = []
for sr in np.logspace(-3, -0.3, 20):
    sr_params.append({'success_rate': sr, 'uplift': uplift, 'beta': 500, 'skew': skew, 'N': N, 'NN': NN})
```

# Data Generation

```python
for param in tqdm(beta_params + skew_params + sr_params):
    apply_all_tests('../data', **param)
```

# Formulas

$$M = \frac{1}{|Users|} \sum_{u \in Users} clicks_u$$

$$H_0: M_{control} = M_{treatment}$$
$$H_1: M_{control} \neq M_{treatment}$$

$$successRate_{treatment} = successRate_{control} (1 + uplift)$$

$$globalCTR = \frac{\sum_{u \in Users} clicks_u}{\sum_{u \in Users} views_u}$$

$$globalCTR = \frac{\sum_{u \in Users} clicks_u}{\sum_{u \in Users} views_u} = \frac{\sum_{u \in Users} views_u * userCTR_u}{\sum_{u \in Users} views_u} $$

$$clicks = k * views$$

$$linearizedClicks = clicks - k * views $$

$$smoothedCTR_u = \frac{clicks_u + \alpha * globalCTR}{views_u + \alpha}$$

$$w_u = \frac{views_u}{1 + \rho(views_u - 1)}$$
